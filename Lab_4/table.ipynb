```python
import numpy as np
def u(x, y):
    return np.exp(1 - x**2 - y**2)

n = 20
m = 20
a, b, c, d = -1, 1, -1, 1
h = (b - a)/n
k = (d - c)/m
xi = [a + i*h for i in range(0, n + 1)]
yj = [c + j*k for j in range(0, m + 1)]
U = []

for i in range(0, m + 1):
    U.append([])
    for j in range(0, int(n/2) + 1):
        U[i].append(0)
        U[i][j] = u(xi[j], yj[i])

for i in range(int(m/2), m + 1):
    for j in range(int(n/2) + 1, n + 1):
        U[i].append(0)
        U[i][j] = u(xi[j], yj[i])

data = pd.DataFrame(U)
data
```
