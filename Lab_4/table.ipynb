```python
import matplotlib.pyplot as plt
D, V_ = [], []
for i in range(0, 10):
    D.append(i/10)
    V_.append(0.957791 + 0.133568*D[i] - 0.2284091*D[i]**2)

V = [0.957, 0.969, 0.976, 0.978, 0.975, 0.968, 0.954, 0.939, 0.918, 0.894]

sum1 = 0
for i in range(0, 10):
    sum1 += (V_[i] - 0.9528)**2
sum2 = 0
for i in range(0, 10):
    sum2 += (V[i] - 0.9528)**2
R2 = sum1/sum2
print('R2', R2)

sum3 = 0
for i in range(0, 10):
    sum3 += (V_[i] - V[i])**2
S = sum3/7
print('S', S)
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
def func(a):
    for x in range(len(a)):
        plt.plot(a[x].real, a[x].imag, 'o-', label='python')
    limit=np.max(np.ceil(np.absolute(a))) # set limits for axis
    plt.xlim((-limit,limit))
    plt.ylim((-limit,limit))
    plt.ylabel('Imaginary')
    plt.xlabel('Real')
    plt.show()
 
omega = np.linspace(0, 1, 2)
a = 1j*(omega**5) + 4*(omega)**4 - 5j*(omega)**3 - 6j*(omega) + 8       # Меняется чтобы рисовать годограф
func(a)
```
