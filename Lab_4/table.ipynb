```python
from sympy import *

# create a "symbol" called x
x, y = symbols('x y')

#Define function
f = exp(1 - x**2 - y**2)

#Calculating Derivative
derivative_f = diff(diff(diff(diff(f, x), x), x), x)

derivative_f
```

```python
derivative_f = diff(diff(diff(diff(f, y), y), y), y)
derivative_f
```
